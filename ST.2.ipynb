```python
import numpy as np
import matplotlib.pyplot as plt

est = []
true = 1 / 32 + np.linspace(-1, 1, 1000) ** 2 / 4

X = np.random.normal(size=(1000, 8))

for theta in np.linspace(-1, 1, 1000):
    est.append(np.mean(((X + theta).mean(axis=-1) / 2 - theta) ** 2))

plt.plot(np.linspace(-1, 1, 1000), est, label='estimated')
plt.plot(np.linspace(-1, 1, 1000), true, label='true')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(n_jobs=-1)
X = np.array([3.49e15, 3.22e15, 3.1e15, 2.95e15, 2.82e15]).reshape(-1, 1)
Y = [0.92, 0.78, 0.68, 0.58, 0.5]
model.fit(X, Y)
model.coef_
```
