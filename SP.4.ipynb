Нужно для каждой выборки построить график функции правдоподобия.

*a).* Параметрическая модель $\mathcal{N}(\theta, 1)$, выборки: `[[-1, 1], [-5, 5], [-1, 5]]`

*b).* Параметрическая модель $Exp(\theta)$, выборки: `[[1, 2], [0.1, 1], [1, 10]]`

*c).* Параметрическая модель $U[0, \theta]$, выборки: `[[0.2, 0.8], [0.5, 1], [0.5, 1.3]]`

*d).* Параметрическая модель $Bin(5, \theta)$, выборки: `[[0, 1], [5, 5], [0, 5]]`

*e).* Параметрическая модель $Pois(\theta)$, выборки: `[[0, 1], [0, 10], [5, 10]]`

*f).* Параметрическая модель $Сauchy(\theta)$, где $\theta$ — параметр сдвига, выборки: `[[-0.5, 0.5], [-2, 2], [-4, 0, 4]]`

Выполнить задание, не создавая много кода, поможет следующая функция.

```python
def draw_likelihood(density_function, grid, samples, label):
    '''
    Изображает график функции правдоподобия для каждой из 3 выборок.
    
    Аргументы:
        density_function --- функция, считающая плотность 
            (обычную или дискретную). На вход данная функция 
            должна принимать массив размера (1, len_sample)
            и возвращать массив размера (len_grid, len_sample).
        grid --- массив размера (len_grid, 1), являющийся 
                 сеткой для построения графика;
        samples --- три выборки;
        label --- latex-код параметрической модели.
    '''
    
    assert len(samples) == 3, "Число выборок не равно 3."

    plt.figure(figsize=(18, 5))
    for i, sample in enumerate(samples):
        sample = np.array(sample)[np.newaxis, :]
        likelihood = np.prod(density_function(sample), axis=1)

        plt.subplot(1, 3, i+1)
        plt.plot(grid, likelihood)
        plt.xlabel('$\\theta$', fontsize=16)
        plt.grid(ls=':')
        plt.title(label + ', sample=' + str(sample), fontsize=16)
    plt.show()
```

Первый пункт можно выполнить с помощью следующего кода:

```python
grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.norm(loc=grid).pdf, grid, 
                [[-1, 1], [-5, 5], [-1, 5]], '$\\mathcal{N}(\\theta, 1)$')
```

Таким образом, распределение времени работы сервера до поломки обладает свойством отсутствия памяти. Как вы знаете, таким свойством среди непрерывных распределений обладает только экспоненциальное распределение.

Компания начала проводить эксперимент, но сервера получились настолько хорошими, что многие еще не сломались, а бизнес уже требует запускать их продажи. Тем самым усложняется задача для аналитиков, которым приходит "обрезанная" выборка.

Рассмотрим задачу более формально. Пусть $X = (X_1, \ldots, X_n)$ &mdash; выборка из экспоненциального распределения с параметром $\theta$ с плотностью $p_\theta(x) = \theta e^{-\theta x}I\{x \geq 0\}$. \
При этом наблюдается выборка $Y = (Y_1, \ldots, Y_n)$, где, $Y_i$ определены следующим образом:
$$
\begin{equation}
  Y_i=\begin{cases}
    X_i, & \text{если $X_i \leq c$}.\\
    c, & \text{если $X_i > c$}.
  \end{cases}
\end{equation}
$$
Иначе говоря, $Y_i = \min(X_i, c)$, где $c$ &mdash; некоторая константа.
Выборка является смесью непрерывного и дискретного распределения, поэтому функция правдоподобия не определена. Вам предлагается реализовать и сравнить три подхода к оценке неизвестного параметра:

2. Найти оценку, используя метод моментов для всей выборки $Y$.
3. Найти ОМП по дискретизованной выборке $Y$, выполнив следующие действия.

    а) Разбить полуинтервал $[0, c)$ на $k$ бинов (множеств) вида $[a_i, a_{i+1})$, где $0=a_0<a_1<...<a_k=c$. В качестве границ бинов возьмите целые числа.
    
    б) Определить $Z_i$ как номер бина, в который попадает $Y_i$, считая при этом множество $\{c\}$ как $(k+1)$-й бин. Вам может помочь функция [`np.digitize`](https://numpy.org/doc/stable/reference/generated/numpy.digitize.html).

**Решение:**

```python
theta, c = 0.15, 10
X = sps.expon(scale=1 / theta).rvs(1000)

Y = np.copy(X)
Y[Y > c] = c
```

Для нашего случая нужно рассчитать плотность условного распределения, которая определяется как:

$$
f_\theta(y)=\frac{p_\theta(y)}{F_\theta(c)}=\frac{p_\theta(y)}{1-e^{-\theta c}}
$$

$$
l_Y(\theta)=-n\log (1-e^{-\theta c})+n\log \theta-\theta\sum_{i=1}^n Y_i
$$

$$
l_Y'(\theta)=-\frac{cn}{e^{\theta c}-1}+\frac{n}{\theta}-\sum_{i=1}^n Y_i
$$

Далее произведём численный подсчёт ОМП.

```python
grid = np.linspace(0.0001, 10, 9999)
est_mle = []
for k in range(1, 1001):
    Y_low = Y[:k][Y[:k] < c]
    dloglike = -c*Y_low.shape[0] / (-1 + np.exp(c * grid)) + Y_low.shape[0] / grid - np.sum(Y_low)
    est_mle.append(grid[np.argmin(np.abs(dloglike))])
print('theta = %.3f' % est_mle[-1])
```

Для метода моментов вычислим математическое ожидание:

$$
\mathsf{E} Y = \int_0^c tp_\theta(t)dt+c\int_c^{\infty} p_\theta(t)dt=\frac{1}{\theta}-\frac{\theta c+1}{\theta}e^{-\theta c}+c e^{-\theta c}=\frac{1-e^{-\theta c}}{\theta}
$$

```python
est_mom = []
for k in range(1, 1001):
    expect = np.mean(Y[:k])
    moment = (1 - np.exp(-grid * c)) / grid
    est_mom.append(grid[np.argmin(np.abs(moment - expect))])
print('theta = %.3f' % est_mom[-1])
```

Теперь рассмотрим ОМП по дискретизированной выборке.

Вычислим вероятность попадания в каждый бин:

$$
p_j(\theta)=\int_{a_j}^{a_{j+1}} p_\theta(t)dt=\int_{a_j}^{a_{j+1}} \theta e^{-\theta t}dt=e^{-\theta a_j}-e^{-\theta a_{j+1}}
$$

$$
p_{k+1}(\theta)=\int_c^\infty p_j(\theta)dt=e^{-c\theta}

Функция правдоподобия:

$$
L_Z(\theta)=\prod_{i=1}^n (e^{-\theta Z_i}-e^{-\theta (Z_i+1)})^{\mathbb{I}\{Z_i\neq 11\}}e^{-c\theta \mathbb{I}\{Z_i = 11\}}
$$

$$
l_Z(\theta)=-c\theta\sum_{i=1}^n \mathbb{I}\{Z_i=11\}-\theta\sum_{i=1}^n Z_i\mathbb{I}\{Z_i\neq 11\}+\log(1-e^{-\theta})\sum_{i=1}^n \mathbb{I}\{Z_i\neq 11\}
$$

$$
l_Z'(\theta)=-c\sum_{i=1}^n \mathbb{I}\{Z_i=11\}-\sum_{i=1}^n Z_i\mathbb{I}\{Z_i\neq 11\}+\frac{e^{-\theta}}{1-e^{-\theta}}\sum_{i=1}^n \mathbb{I}\{Z_i\neq 11\}
$$

```python
est_disc = []
bins = np.linspace(0, 10, 11)
for k in range(1, 1001):
    Z = np.digitize(Y[:k], bins)

    Z_low = Z[Z != 11]
    Z_high = Z[Z == 11]

    dloglike_Z = -c * Z_high.shape[0] - np.sum(Z_low) + np.exp(-grid) / (1 - np.exp(-grid)) * Z_low.shape[0]
    est_disc.append(grid[np.argmin(np.abs(dloglike_Z))])
print('theta = %.3f' % est_disc[-1])
```

Ранее мы выводили оценки для всей выборки, теперь построим графики оценок в зависимости от размера выборки.

```python
n_range = range(1, 1001)
plt.figure(figsize=(12, 6))

plt.plot(n_range, est_mle, label='ОМП')
plt.plot(n_range, est_mom, label='Метод моментов')
plt.plot(n_range, est_disc, label='Дискретизация')
plt.hlines(0.15, 0, 1000, color='red', lw=3, alpha=0.5, label='Истинное значение')

plt.xlabel('Размер выборки')
plt.ylabel('$\\hat{\\theta}$')
plt.title('Зависимость значений оценок от размера выборки')
plt.legend()
plt.ylim((0.05, 0.25))
plt.show()
```

**Вывод:** Наилучший результат дал метод моментов, так как математическое ожидание мы можем оценить для любой выборки, и оно однозначно зависит от $\theta$. Оценка же максимального правдоподобия в первом случае отбрасывает всю информацию об элементах больше $c$, а во втором - приводит такие элементы, как будто они все оказались в пределах $[10;11)$. В итоге оценки получаются отклоняющимися от действительности. Оценка оказывается тем точнее, чем больше элементов в выборке, причём начиная от выборки в примерно 200 элементов значения оценок помимо ОМП изменяются мало, в пределах $\pm 10^{-2}$. ОМП же "стабилизируется" после примерно 700 элементов, что закономерно - ОМП строится по подвыборке.
